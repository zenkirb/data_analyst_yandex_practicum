### Предоставленные данные

В нашем распоряжении исторические данные о продажах игр по 2016 год по всем платформам и регионам, жанре игр и оценкам пользователей и экспертов.

Описание данных таблицы:

* `Name` — название игры
* `Platform` — платформа
* `Year_of_Release` — год выпуска
* `Genre` — жанр игры
* `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
* `EU_sales` — продажи в Европе (миллионы проданных копий)
* `JP_sales` — продажи в Японии (миллионы проданных копий)
* `Other_sales` — продажи в других странах (миллионы проданных копий)
* `Critic_Score` — оценка критиков (максимум 100)
* `User_Score` — оценка пользователей (максимум 10)
* `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию.

Расшифровка рейтинга ESRB по данным из [Википедии](https://ru.wikipedia.org/wiki/Entertainment_Software_Rating_Board):

* `EC` («Early childhood») — «Для детей младшего возраста»: Игра подходит для детей от 3 лет и не содержит материалов, которые родители могли бы счесть неподходящими. Продукты, получившие данный рейтинг, изначально разрабатываются для детей и обычно представляют собой развивающие игры. Некоторые усложнённые развивающие игры могут иметь рейтинг «Everyone».
* `E` («Everyone») — «Для всех»: Содержание вполне подходит для возрастной категории от 6 лет. Такие игры могут понравиться и взрослым. Игры с этим рейтингом могут содержать минимальное насилие, в основном «мультяшного» характера.
* `E10+` («Everyone 10 and older») — «Для всех от 10 лет и старше»: Проекты с данным рейтингом могут содержать или немного больше мультипликационного или мягкого насилия, или несколько откровенные сцены, или минимальное количество крови.
* `T` («Teen») — «Подросткам»: Игра подходит для лиц от 13 лет. Проекты из данной категории могут содержать насилие, непристойные сцены, грубый юмор, в меру откровенное сексуальное содержимое, кровь или нечастое использование ненормативной лексики.
* `M` («Mature») — «Для взрослых»: Материалы игры не подходят для подростков младше 17 лет. Проекты с данным рейтингом могут содержать достаточно жестокое насилие, большое количество крови с расчленением, непристойные сексуальные сцены или грубую ненормативную лексику, нежелательную для младшей аудитории.
* `AO` («Adults Only 18+») — «Только для взрослых»: Содержание игры только для взрослых старше 18 лет. Продукты из данной категории могут содержать длительные сцены жестокого насилия и/или очень откровенное сексуальное содержимое, а также сцены с обнажением.
* `RP` («Rating Pending») — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоения рейтинга. Данный логотип используется только на рекламных презентациях и в демо-версиях игр до официальной даты выпуска в продажу.

Так же, нас сразу предупреждают, что данные за 2016 год могут быть неполными.

```python
# напишем цикл, который будет выводить диаграмму рассеивания по каждой платформе по отзывам критиков и
# пользователей к глобальным продажам. И выведем коэффициенты корреляции по таким же параметрам.

for elem in actual_top['platform'].unique():
        actual_top.query('platform == @elem').plot(kind='scatter',
                                                   x='critic_score',
                                                   y='total_sales',
                                                   figsize=(14,4))
        plt.title(f'Диаграмма рассеивания {elem} по оценкам критиков к продажам')
        plt.xlabel('Оценка критиков')
        plt.ylabel('Продажи, млн')
        plt.show()
        
        
        actual_top.query('platform == @elem').plot(kind='scatter',
                                                   x='user_score',
                                                   y='total_sales',
                                                   color='orange',
                                                   figsize=(14,4))
        plt.title(f'Диаграмма рассеивания {elem} по оценкам пользователей к продажам')
        plt.xlabel('Оценка пользователей')
        plt.ylabel('Продажи, млн')
        plt.show()
        
        
        print(f'{elem} коэф.корреляции оценок критиков к продажам: {actual_top.query("platform == @elem")["critic_score"].corr(actual_top.query("platform == @elem")["total_sales"]):.2f}')
        print(f'{elem} коэф.корреляции оценок пользователей к продажам: {actual_top.query("platform == @elem")["user_score"].corr(actual_top.query("platform == @elem")["total_sales"]):.2f}')
        
        print()
```

### Проверка гипотез

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Проверим гипотезу, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Нулевая гипотеза - средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
Альтернативная гипотеза - средние пользовательские рейтинги платформ Xbox One и PC различаются.

```python
# создадим два списка с рейтингами платформ XOne и PC за всё время, отсеяв пропущенные значения
xboxone_rating_all = data.query('(user_score >= 0) & (platform == "XOne")')['user_score']
pc_rating_all = data.query('(user_score >= 0) & (platform == "PC")')['user_score']

# и два списка с рейтингами платформ XOne и PC за актуальный период
xboxone_rating_actual = actual_data.query('(user_score >= 0) & (platform == "XOne")')['user_score']
pc_rating_actual = actual_data.query('(user_score >= 0) & (platform == "PC")')['user_score']
```

```python
# проведем проверку гипотезы по рейтингам платформ XOne и PC за всё время,
# уровень статистической значимости укажем 5%, а равность дисперсий укажем False, 
# так как мы не знаем, что наши выборки равны 

results = st.ttest_ind(xboxone_rating_all,
                       pc_rating_all,
                       equal_var = False)

alpha = 0.05

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
# выведем средние оценки на платформах Xbox One и PC за всё время
print("Средняя оценка пользователя на платформе Xbox One: {:.2f}"
      .format(xboxone_rating_all.mean()))

print("Средняя оценка пользователя на платформе PC: {:.2f}"
      .format(pc_rating_all.mean()))
```

```python
# проведем проверку гипотезы по рейтингам платформ XOne и PC за актуальный период,
# уровень статистической значимости укажем 5%, а равность дисперсий укажем False, 
# так как мы не знаем, что наши выборки равны 


results = st.ttest_ind(xboxone_rating_actual,
                       pc_rating_actual,
                       equal_var = False)

alpha = 0.05

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
# выведем средние оценки на платформах Xbox One и PC за актуальный период
print("Средняя оценка пользователя на платформе Xbox One: {:.2f}"
      .format(xboxone_rating_actual.mean()))

print("Средняя оценка пользователя на платформе PC: {:.2f}"
      .format(pc_rating_actual.mean()))
```

#### Вывод

Проведя проверку гипотезы о том что средние пользовательские рейтинги платформ Xbox One и PC одинаковые **взяв данные за всё время** - **нулевая гипотеза была отвергнута**, а значит средние пользовательские рейтинги не равны. Однако, платформа `PC` существует уже 31 год, а платформа `XOne` существует всего 3 года (на момент конца 2016 года), следовательно данные по платформе `PC` будут слишком искальжены по отношению к платформе `XOne` и поэтому мы проверили дополнительную проверку гипотезы по актуальному периоду (20012-2016 года).

Проверка гипотезы на основе **данных по актуальному периоду** показала, что **отвергнуть нулевую гипотезу не получилось**, а значит средние пользовательские рейтинги двух платформ равны. Значение P-value равняется ~55%, что говорит нам о том, что данный результат можно получить случайно в ~55% случаев, при условии, что нулевая гипотеза верна.
