### Категоризация клиентов

Разобъем клиентов по категориям скоринга, возраста и зарплат и изучим их.

#### Категории клиентов по кредитному скорингу

На основе распределения значений в столбце `score`, разобъем клиентов на три группы по уровню кол-ва баллов кредитного скоринга:
* "Низкий уровень" - клиенты с количеством баллов ниже 500
* "Средний уровень" - клиенты с количеством баллов от 500 до 700
* "Высокий уровень" - клиенты с количеством баллов выше 700

```python
# создадим функцию, которая проверяет кол-во баллов клиента в нашем датафрейме и возвращает категорию
def score_category(df):
    if df['score'] <= 500:
        return 'Низкий уровень'
    elif (df['score'] >= 501 and df['score'] <= 700):
        return 'Средний уровень'
    else:
        return 'Высокий уровень'
```

```python
# создадим новый столбец и применим функцию к датафрейму и выведем результат
df['score_category'] = df.apply(score_category, axis=1)
df
```

#### Категории клиентов по возрасту

Разобъем клиентов по стандартным категориям возраста:
* 18-45 лет
* 46-60 лет
* старше 60 лет (пенсионный возраст)

```python
# создадим функцию, которая проверяет возраст клиента в нашем датафрейме и возвращает категорию
def age_cat(df):
    if (df['age'] >= 18 and df['age'] <= 45):
        return '18-45'
    elif (df['age'] > 45 and df['age'] <= 60):
        return '46-60'
    else:
        return '61+'
```

```python
# создадим новый столбец и применим функцию к датафрейму и выведем результат
df['age_category'] = df.apply(age_cat, axis=1)
df
```

#### Категории клиентов по доходу

Воспользуемся распределением зарплат из нашего датасета и разобъем клиентов по трем категориям дохода: 
* "Низкий доход" - клиенты с зарплатой до 50 тыс. 
* "Средний доход" - клиенты с зарплатой 50-100 тыс.
* "Высокий доход" - клиенты с зарплатой свыше 100 тыс.

```python
# создадим функцию, которая проверяет значение зарплаты клиента в нашем датафрейме и возвращает категорию
def salary_cat(df):
    if df['estimated_salary'] < 50000:
        return 'Низкий доход'
    
    elif (df['estimated_salary'] >= 50000 and df['estimated_salary'] <= 100000):
        return 'Средний доход'
    
    else:
        return 'Высокий доход'
```

### Проверка гипотез

```python
# создадим два датафрейма с клиентами в оттоке и оставшимися 
churn_1 = df.query('churn == 1')
churn_0 = df.query('churn == 0')
```

```python
churn_0
```

#### Проверка гипотезы о различии доходов

Проверим гипотезу о различии дохода между теми клиентами, которые ушли и теми, которые остались.

```python
# посчитаем дисперсию и средние значения зарплат клиентов в оттоке и оставшихся
var_churn_1 = np.array(churn_1['estimated_salary'])
var_churn_0 = np.array(churn_0['estimated_salary'])

print('Дисперсия зарплат оставшихся клиентов:', np.var(var_churn_0))
print('Дисперсия зарплат ушедших клиетов:', np.var(var_churn_1))
print()
print('Средняя зарплата оставшихся клиентов:', np.mean(var_churn_0))
print('Средняя зарплата ушедших клиетов:', np.mean(var_churn_1))
```

Сформируем нулевую и альтернативную гипотезы:

* Нулевая гипотеза - доходы между теми, кто ушел в отток и теми, кто остался равны
* Альтернативная гипотеза - доходы между теми, кто ушел в отток и теми, кто остался не равны

Так как данные по доходу у нас имеют нормальное распределение и дисперсии практически равны, то мы будем проверять гипотезы используя t-критерий Стьюдента.

```python
# проведем t-test 
alpha = .05
results = st.ttest_ind(var_churn_1, var_churn_0)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

##### Вывод

По результату t-теста мы можем сделать вывод, что различия в доходах между оставшихся и ушедших клиентах не имеют статистической значимости и значит, что мы не можем отвергнуть нулевую гипотезу. Следовательно доходы между группами равны

#### Проверка гипотезы о различии в возрасте

Проверим гипотезу о различии возраста среди ушедших клиентов и оставшихся.

```python
# посчитаем дисперсию и средние значения возрастов клиентов в оттоке и оставшихся
age_churn_0 = np.array(churn_0['age'])
age_churn_1 = np.array(churn_1['age'])

print('Дисперсия возраста оставшихся клиентов:', np.var(age_churn_0))
print('Дисперсия возраста ушедших клиетов:', np.var(age_churn_1))
print()
print('Средний возраст оставшихся клиентов:', np.mean(age_churn_0))
print('Средний возраст ушедших клиетов:', np.mean(age_churn_1))
```

```python
# выведем график распределения значений возраста у ушедших и оставшихся клиентов
plt.figure(figsize=(15,8))
sns.distplot(churn_1['age'],label='Ушли')
sns.distplot(churn_0['age'],label='Остались')
plt.title('Распределение возраста клиентов')
plt.legend()
plt.show()
```

Сформируем нулевую и альтернативную гипотезы:
* Нулевая гипотеза - возраст между теми, кто ушел в отток и теми, кто остался равны
* Альтернативная гипотеза - возраст между теми, кто ушел в отток и теми, кто остался не равны

Так как данные по возрасту у нас не имеют нормального распределение и дисперсии не равны, то мы будем проверять гипотезы используя t-критерий Стьюдента с указанием equal_var=False.

```python
# проведем t-test 
alpha = .05
results = st.ttest_ind(age_churn_1, age_churn_0, equal_var=False)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

##### Вывод

По критерию Стьюдента значение p-value меньше уровня статистической значимости. Следовательно средний возраст ушедших и оставшихся клиентов действительно различается, а значит нулевая гипотеза отвергается. Это так же видно на графике и на средних значениях возраста клиентов, ушедших в отток (44 года) и оставшихся (37 лет).
