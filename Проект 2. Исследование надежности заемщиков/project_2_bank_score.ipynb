# Исследование надежности заемщиков.

Кредитному отделу банка, для построения модели кредитного скоринга, необходимо узнать влияет ли семейное положение и количество детей у клиента банка на погашение им кредитов в срок. Также, нужно определить, есть ли другие зависимости на основе статистике по платежёспособности.

**Данные для исследования** - статистика о платёжеспособности клиентов.

**Цель исследования** - определить, влияют ли на погашение кредита в срок следующие факторы:
1. Семейное положение клиента
2. Количество детей у клиента

**Дополнительные гипотезы**, установленные в ходе исследования: 
1. Есть ли зависимость между уровень дохода клиента и погашением кредита в срок
2. Есль ли зависимость между целью кредита и погашением кредита в срок
3. Есть ли зависимость между уровнем образования и погашением кредита в срок
4. Если ли зависимость между должностью и погашением кредита в срок
5. Есть ли зависмость между полом клиента и погашением кредита в срок

**Ход исследования:**
1. Изучение полученных данных
2. Предобработка данных
3. Проверка гипотез и выводы
4. Итоги исследования

## Шаг 2. Предобработка данных

### Шаг 2.0. Удаление явных дубликатов

Перед проверкой данных на аномалии и заполнением пропусков, сперва удалим явные дубликаты.

```python
# удаляем явные дубликаты в таблице с перезаписью индексов и без создания нового столбца со старыми индексами
print('Количество явных дубликатов до:', df.duplicated().sum())
df = df.drop_duplicates().reset_index(drop=True)
print('Количество явных дубликатов после:', df.duplicated().sum())
```

### Шаг 2.1. Проверка и исправления данных на аномалии

На этом этапе нам необходимо разобраться с аномалиями количественных значениях в столбцах и исправить их.

Начнем со столбца `children`, в котором мы обнаружили несколько аномальных значений.

```python
# выведем на экран уникальные значение столбца 'children'
df['children'].value_counts()
```

Исходя из полученных значений, мы можем предположить, что значения "20" и "-1" появились в результате ошибки ввода данных. Исправим значение "20" на "2", а "-1" исправим на "0". Так как вероятнее всего значение "-1" обозначает пропуск или нулевое значение, а количества этих значений достаточно малы, поэтому данное исправление не будет сильно искажать данные.

```python
# меняем значение 20 в столбце 'children' на цифру 2
# меняем значение -1 в столбце 'children' на цифру 1
# и выводим результат
df['children'] = df['children'].replace(20, 2)
df['children'] = df['children'].replace(-1, 0)
df['children'].value_counts()
```

Теперь нам нужно разобраться со столбцом `days_employed` убрать аномалии и привести все значения в один вид и далее убрать пропуски.

```python
# изучим количества значений в столбце 'days_employed'
print('Количество отрицательных значений:', df.loc[df['days_employed'] < 0, 'days_employed'].count())
print('Количество положительных значений:', df.loc[df['days_employed'] > 0, 'days_employed'].count())
print('Количество нулевых значений:', df.loc[df['days_employed'] == 0, 'days_employed'].count())
print('Количество пропущенных значений:', df['days_employed'].isna().sum())
```

Исходя из того, что в столбце `days_employed` хранятся значения количества дней - нам необходимо поменять отрицательные значение на положительные и затем изучить данных на аномалии.

```python
# возводим все отрицательные значение модуль и проверяем результат
df['days_employed'] = abs(df['days_employed'])
print('Количество отрицательных значений:', df.loc[df['days_employed'] < 0, 'days_employed'].count())
print('Количество положительных значений:', df.loc[df['days_employed'] > 0, 'days_employed'].count())
print('Количество нулевых значений:', df.loc[df['days_employed'] == 0, 'days_employed'].count())
```

Теперь необходимо изучить значения столбца `days_employed` на аномальные значения.

```python
# смотрим максимальное и минимальное значение в столбце 'days_employed'
print('Максимальное значение:', df['days_employed'].max())
print('Минимальное значение:', df['days_employed'].min())
```

```python
# выведем на экран строку с максимальным значением
df[df['days_employed'] == df['days_employed'].max()]
```

```python
# посмотрим max значение дней в года
df['days_employed'].max() / 365
```

1100 лет трудового стажа. Либо перед нами трудолюбивый эльф, скрывающий свой реальный возраст, либо перед нами ошибка в значениях.
Посмотрим сколько "эльфов" у нас скрывается в данных. Для этого мы предположим, что максимальный трудовой стаж в нашем исследовании может быть у мужчины пенсионера 65 лет, который начал работать в 18 лет. И сравним разницу между максимально возможным рабочим стажем исходя из максимального возраста, который есть в наших данных.

```python
# найдем количество рабочих лет, отняв 18 лет от возраста и умножим на количество дней в году
max_work_days = (65 - 18) * 365
max_work_days_in_dob = (df['dob_years'].max() - 18) * 365
print('Максимальное количество отработанных дней в теории:', max_work_days)
print('Максимальное количество отработанных дней по max возрасту:', max_work_days_in_dob)
```

```python
# выведем на экран количество строк, которые имеют значения больше чем max кол-во отработанных дней
print('Количество "эльфов" в теории:', df.loc[df['days_employed'] >= max_work_days, 'days_employed'].count())
print('Количество "эльфов" по max возрасту:', df.loc[df['days_employed'] >= max_work_days_in_dob, 'days_employed'].count())
```

Количество аномальных значений отфильтрованных по теоретическому максимальному рабочему стажему почти одинаково, но переменная с теоретическим 65 летним пенсионеров показала результат лучше. Эти переменную и возьмем за основную.

Предположим, что аномальные значения рабочих дней указаны в часах и проверим эту теорию.

```python
# выведем список из 10 значений из столбца 'days_employed' отфильрованного по max_work_days
# переведя значения в часы, а затем в года и сохранив в переменной max_work_years
max_work_years = df.loc[df['days_employed'] >= max_work_days, 'days_employed'] / 24 / 365
max_work_years.head(10)
```

38-45 лет рабочего стажа - наша теория подтвердилась. Ошибка в значениях появилась из-за перевода значений в часы, вместо суток.
Теперь можно исправить ошибочные данные в столбце `days_employed`, переведя все аномальные значения из часов в сутки.

```python
# заменим аномальные значения в таблице df, разделив аномальные значения на 24 часа
# и выведем на экран список из 10 строк таблицы, отсортированных по максимальному значению 
df.loc[df['days_employed'] >= max_work_days, 'days_employed'] = ( 
    df.loc[df['days_employed'] >= max_work_days, 'days_employed'] / 24 
)
df.sort_values(by='days_employed', ascending=False).head(10)
```

С аномальными значениями в столбце `days_employed` вопрос решен. Останется только заполнить пустные значения. Этот пункт мы сделаем в следующем шаге, а сейчас нам нужно убрать аномальное значение в столбце `dob_years`.

При изучении данных, в столбце `dob_years` мы обнаружили, что в 101 строке указано значение "0". Так как в этом столбце указано значение возраста клиентов - нулевое значение быть не может. Так как это числовое значение, то мы заменим аномалию на медианное значение по столбцу.

```python
# выведем на экран медианное значение возраста
df['dob_years'].median()
```

```python
# заменим нулевые значение в столбце 'dob_years' на медианное значение
print('Количество нулевых значений до:', df.loc[df['dob_years'] == 0, 'dob_years'].count())
df.loc[df['dob_years'] == 0, 'dob_years'] = df['dob_years'].median()
print('Количество нулевых значений после:', df.loc[df['dob_years'] == 0, 'dob_years'].count())
```

Нулевые значения в столбце `dob_years` убраны.

Теперь переходим к аномалии в столбце `gender` - обнаружена строка с неопределенном полом клиента.

```python
# выведем на экран все значения столбца 'gender'
df['gender'].value_counts()
```

Исходя из того, что неопределенное значение есть только в единственной строке таблицы - можно удалить строку, так как это не повлияет на итоговые выводы исследования.

```python
# выведем на экран все уникальные значения и их количества столбца 'purpose'
df['purpose'].value_counts()
```

Большое количество похожих по смыслу значений, которые явно появились из-за ручного ввода.
Необходимо привести всё к 4-ем категориям:
* Операции с автомобилем
* Операции с недвижимостью
* Проведение свадьбы
* Получение образования

```python
# создадим функцию, которая будет проверять значения на совпадения по словарю и заменять их
def transform_purpose(element):
    purpose_dict = {
        'операции с автомобилем' : ['авто'],
        'операции с недвижимостью' : ['жил', 'недвиж'],
        'проведение свадьбы' : ['свадьб'],
        'получение образования' : ['образ']
    }
    for category, synonymns in purpose_dict.items():
        for syn in synonymns:
            if syn in element:
                return category
    return None
```

```python
# создадим новый столбец 'purpose_category' в таблице со значениями категорий и выведем значения и их кол-во
df['purpose_category'] = df['purpose'].apply(transform_purpose)
df['purpose_category'].value_counts()
```

### Шаг 2.6. Категоризация дохода.

Теперь создадим отдельный словарь и столбец по категориям дохода.
Для более удобного сравнения клиентов по доходу, разделим доходы на следующие категории:
* Доход от 0 до 30000 - категория 'E'
* Доход от 30001 до 50000 —  категория 'D'
* Доход от 50001 до 200000 — категория 'C'
* Доход от 200001 до 1000000 — категория 'B'
* Доход от 1000001 и выше — категория 'A'

```python
# создадим функцию, которая возвращает значение категории в зависимости от значения дохода
def ins_income_category(income):
    if income <= 30000:
        return 'E'
    if income >= 30001 and income <= 50000:
        return 'D'
    if income >= 50001 and income <= 200000:
        return 'C'
    if income >= 200001 and income <= 1000000:
        return 'B'
    if income >= 1000001:
        return 'A'
```

### Шаг 3.0. Подготовка к проверке

Как мы знаем, столбец `debt` в нашей таблице содержит значения '0' и '1', что означает, что у клиента нету долга либо есть долг соответственно.
Чтобы понять, есть ли зависимость между разными критерями и возвратом долга клиентом, мы проверим какую долю составляют невозвраты.

```python
# напишем функцию, которая будет создавать сводную таблицу на основне таблицы и нужного столбца с критерями
# затем высчитывать долю задолженностей и выводить отсортированный по доле результат.
# основные колонки 'debt' будут переименованы и оставим возможность, при необходимости
# менять столбец столбец значений, функций и сортировки.
# столбец значения оставим по умолчанию, например, 'purpose',так как нам нужен только подсчет количества значений.
def debt_pivot_table(data, pivot_index, 
        pivot_columns='debt' ,pivot_values='purpose', pivot_aggfunc='count', sort='Доля долгов'):
    name = data.pivot_table(
        index=[pivot_index],
        columns=[pivot_columns],
        values=[pivot_values],
        aggfunc=pivot_aggfunc
    ).fillna(0) # меняем пустые значение на нули, если они появятся
    name.columns = ['Нет долгов',  'Есть долги']
    name['Доля долгов'] = name['Есть долги'] / (name['Есть долги'] + name['Нет долгов'])
    return name.sort_values(by=sort, ascending=False)
```

```python
# функция которая будет принимать сводную таблицу и форматировать колонки столбцов долгов,
# и затем выводить на экран результат
def debt_style_format(table_pivot):
    return display(
        table_pivot.style.format({'Нет долгов':'{:.0f}', 'Есть долги':'{:.0f}','Доля долгов':'{:.2%}'})
                  )
```

Функции готовы, теперь можем с помощью функции  `debt_pivot_table` вывести нужные нам сводные таблицы и ответить на вопросы.
