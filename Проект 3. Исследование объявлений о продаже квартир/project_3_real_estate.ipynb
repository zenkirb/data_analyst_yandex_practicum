Всего 24 строки с большими значениями. Можно предположить, что это всё таки ошибка ввода, так как, если мы разделим значения на 10, то получим значения, которые будут похожи на стандартную высоту потолков в квартирах.

```python
# выберем все строки у которых высота потолков больше 11 и разделим значения на 10
# и выведем количественную информацию по столбцу
df.loc[df['ceiling_height'] > 11, 'ceiling_height'] /= 10
df['ceiling_height'].describe()
```

Незначительно отклонение снизилось почти на единицу. Посмотрим что происходит с минимальными значениями. 

```python
# выведем список квартир с высотой меньше 2 метров
df.query('ceiling_height < 2')
```

Если при 2 метров потолков еще можно жить, то вот как живут в 4 квартирах с указанной высотой потолков менее 2 метров? Предположим, что это тоже ошибка ручного ввода и добавим этимс квартирам 1 метр высоты.

```python
# увеличим высоту потолков на 1 метр у всех квартир, где значение меньше 2 метров
# выведим информацию по столбцу
df.loc[df['ceiling_height'] < 2, 'ceiling_height'] += 1
df['ceiling_height'].describe()
```

```python
# выведем график, чтобы посмотреть как сейчас распределяются значения
df.boxplot(column='ceiling_height')
```

```python
# выведем список квартир у которых высота потолка более 3.5 метров
df.query('ceiling_height > 3.5')
```

Квартир с очень высокими потолками немного и быстрый взгляд показывает, что это большие и дорогие квартиры, которые могут быть и двухэтажными. 
Теперь можно заполнить пропуски в столбце высоты потолка медианными значениями.

```python
# заполним пропуски медианным значением и проверим количество пропусков
# затем выведем информацию по столбцу
print('Количество пропущенных значений до:', df['ceiling_height'].isna().sum())
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
print('Количество пропущенных значений после:', df['ceiling_height'].isna().sum())

df['ceiling_height'].describe()
```

**Вывод**

В столбце `ceiling_height` в котором хранится информация по высоте потолка мы заполнили пропуски медианным значениям, предварительно предположив что аномальные значения появились по причине ошибки ввода данных пользователем и заменив их на более реальные значения.

Столбец `floors_total` хранит значения количество этажей в доме. Выведем информацию по столбцу и пропусках.

```python
# выведем информацию по стобцу
df['floors_total'].describe()
```

```python
# выведем отсортированный список квартир с пропущенным значеним этажей в доме 
df[df['floors_total'].isna()].sort_values(by='floor')
```

Всего 86 строк, что относительно всей таблицы очень мало, а в столбце `floor` пропуски отсутствуют. Поэтому мы можем заполнить пропуски значением этажа квартиры из столбца `floor`, предположив, что квартира находится на последнем этаже.

```python
# выберем все строки с пропусками в столбце 'floors_total' 
# и заменим ячейки с пропусками значениями из столбца 'floor'
print('Количество пропусков до:', df['floors_total'].isna().sum())
df.loc[df['floors_total'].isna(), 'floors_total'] = df.loc[df['floors_total'].isna(), 'floor']
print('Количество пропусков после:', df['floors_total'].isna().sum())
```

```python
# выведем информацию по стобцу
df['floors_total'].describe()
```

```python
# выведем график, чтобы посмотреть как сейчас распределяются значения
df.boxplot(column='floors_total')
```

```python
# посмотрим сколько и какие строки имеют количество этажей больше 30
df.query('floors_total > 30').sort_values(by='floors_total')
```

Строки до 37-ми этажей выглядят реальными, а вот последние две строки с 52 и 60 этажей в здании выглядят уже как ошибки ввода. Особенно интересно выглядит 60 этажное здание в Кронштадте. Поэтому предположим что это явно ошибочные значения и уберем эти строки из таблицы.

```python
# с помощью фильтра уберем в таблице квартиры у которых кол-во этажей в доме 52 и 60
# и выведем информацию по таблице
df = df[df['floors_total'] < 50]
df['floors_total'].describe()
```

### Стоимость квадратного метра

```python
# добавим столбец со стоимостью квадратного метра квартиры и преведем реузультат к целому числу
# выведем результат первых 10 строк
df['square_cost'] = (df['last_price'] / df['total_area']).astype('int')
df.head(10)
```

### Категория этажа квартиры

```python
# напишем функцию, которая будет возвращать значение категории этажа в зависимости от значения этажа
# а затем добавим новый столбец 'floor_category' в который передадим значения функции
# и выведем на экран результат
def floor_category(row):
    if row['floor'] == 1:
        return 'Первый'
    if row['floor'] == row['floors_total']: 
        return 'Последний'
    return 'Другой'

df['floor_category'] = df.apply(floor_category, axis = 1)
df.head(5)
```
